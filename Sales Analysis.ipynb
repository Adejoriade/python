# 3. Find The Distribution of Customer Ratings(Numerical Column)

# Bivariate Analysis/ Multivariate Analysis

# 5. Does The Cost of Goods Sold Affect The Ratings That The Customers Provide? (Numerical-Numerical)

```python
# Scatter plot with regression line

plt.figure(figsize=(15, 6))
sns.regplot(x='cogs', y='Rating', data=df)
plt.title('COGS vs. Customer Ratings')
plt.xlabel('Cost of Goods Sold (COGS)')
plt.ylabel('Customer Rating')
plt.show() 
```

```python
import statsmodels.api as sm
# Calculate the correlation coefficient
correlation = df['cogs'].corr(df['Rating'])
print(f"Correlation coefficient between COGS and Rating: {correlation}")
```

# 6. Does Gross Income Affect The Ratings That The Customers Provide?(Numerical-Numerical)

```python
# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='gross income', y='Rating', data=df)
plt.title('Gross Income vs. Customer Ratings')
plt.xlabel('Gross Income')
plt.ylabel('Customer Rating')
plt.show() 
```

# 7. Find The Most Profitable Branch As Per Gross Income. (Numerical-Categorical)

```python
# Group by branch and sum the gross income
branch_gross_income = df.groupby('Branch')['gross income'].sum().reset_index()

# Sort the branches by gross income in descending order
branch_gross_income = branch_gross_income.sort_values(by='gross income', ascending=False)

# Display the result
print(branch_gross_income)

# Get the most profitable branch
most_profitable_branch = branch_gross_income.iloc[0]
print(f"The most profitable branch is: {most_profitable_branch['Branch']} with a gross income of {most_profitable_branch['gross income']}")

```
